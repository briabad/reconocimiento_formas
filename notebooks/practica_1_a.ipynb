SVD

El Análisis de Valor Singular (Singular Value Decomposition o SVD) es una técnica matemática que se utiliza para reducir la dimensionalidad de un conjunto de datos al descomponer una matriz en tres matrices más pequeñas. La idea detrás del SVD es encontrar una representación de menor dimensión que conserve la mayor cantidad posible de la variabilidad de los datos originales. Matemáticamente, el SVD de una matriz X se realiza de la siguiente manera:

Supongamos que tenemos una matriz X de tamaño m x n que representa nuestros datos, donde m es el número de muestras y n es el número de características (dimensiones).

Descomposición de la Matriz:

X se descompone en tres matrices: U, S, y V^T, donde:
U es una matriz m x m cuyas columnas son autovectores ortonormales de XX^T.
S es una matriz diagonal m x n con valores singulares en la diagonal principal, que representan la importancia de cada dimensión.
V^T es la traspuesta de una matriz n x n cuyas columnas son autovectores ortonormales de X^TX.
Reducción de Dimensiones:

Para reducir la dimensionalidad de los datos, se pueden truncar las matrices U, S, y V^T manteniendo solo las k dimensiones más importantes. Esto se hace seleccionando las k primeras columnas de U, tomando los k primeros valores singulares de S y las k primeras filas de V^T.
La nueva representación de los datos se obtiene como X' = U_k * S_k * (V^T)_k, donde U_k, S_k, y (V^T)_k representan las matrices truncadas.

Regresion logistica

La función de odds (o simplemente "odds") es una medida que se utiliza para representar la probabilidad de un evento en relación con la probabilidad de que el evento no ocurra. En términos matemáticos, la odds de un evento A se define como:

Odds(A) = P(A) / P(¬A)

Donde:

- P(A) es la probabilidad de que el evento A ocurra.
- P(¬A) es la probabilidad de que el evento A no ocurra (es decir, el evento complementario de A).

Las odds pueden tomar cualquier valor no negativo, y un valor de 1 indica que la probabilidad de que ocurra el evento A es igual a la probabilidad de que no ocurra (P(A) = P(¬A)). Si las odds son mayores que 1, significa que el evento A es más probable que ocurra que no ocurra, y si las odds son menores que 1, significa que el evento A es menos probable que ocurra que no ocurra.

En estadísticas y en el contexto de la regresión logística, las odds son una parte fundamental del modelo. En la regresión logística, la relación entre las variables predictoras y la variable de salida se modela en términos de log-odds (logaritmo natural de las odds) de la siguiente manera:

log(P(Y = 1) / P(Y = 0)) = β₀ + β₁x₁ + β₂x₂ + ... + βₙxₙ

Donde:

- P(Y = 1) es la probabilidad de que la variable de salida sea 1.
- P(Y = 0) es la probabilidad de que la variable de salida sea 0.
- β₀, β₁, ... βₙ son los coeficientes del modelo.
- x₁, x₂, ... xₙ son las variables predictoras.

En el contexto de la regresión logística, se utiliza el logaritmo natural para modelar la relación entre las odds de que Y = 1 y las variables predictoras. Las odds y el logaritmo de las odds son conceptos clave en la estadística y en la regresión logística, ya que permiten modelar relaciones entre variables y realizar predicciones.

Para despejar la probabilidad condicional P(Y=1∣X) en términos de la función sigmoide, primero comenzamos con la ecuación de la regresión logística:

log(P(Y = 1) / P(Y = 0)) = β₀ + β₁x₁ + β₂x₂ + ... + βₙxₙ

Para encontrar P(Y=1∣X), aplicamos la función exponencial a ambos lados de la ecuación:

(P(Y = 1) / P(Y = 0)) = e^(β₀ + β₁x₁ + β₂x₂ + ... + βₙxₙ)

Luego, despejamos P(Y=1) dividiendo ambos lados por 1+P(Y=0):

P(Y = 1) = P(Y = 0) * e^(β₀ + β₁x₁ + β₂x₂ + ... + βₙxₙ)

A continuación, sustituimos 1−P(Y=1), ya que estamos tratando con una clasificación binaria donde Y puede ser 0 o 1:

P(Y = 1) = (1 - P(Y = 1)) * e^(β₀ + β₁x₁ + β₂x₂ + ... + βₙxₙ)

Finalmente, resolvemos para P(Y=1) y expresamos la probabilidad condicional en términos de la función sigmoide S(z):

P(Y = 1) = e^(β₀ + β₁x₁ + β₂x₂ + ... + βₙxₙ) / (1 + e^(β₀ + β₁x₁ + β₂x₂ + ... + βₙxₙ)) = 1 / (1 + e^-(β₀ + β₁x₁ + β₂x₂ + ... + βₙxₙ))

Esto es la función sigmoide S(z) que se utiliza en la regresión logística para modelar la probabilidad condicional P(Y=1∣X) en función de las variables predictoras X y los coeficientes del modelo β. La función sigmoide mapea z a un valor en el rango (0, 1) y se utiliza para estimar la probabilidad de que la variable de salida Y sea 1 dadas las características X

La Regresión Logística, el Análisis Discriminante de Fisher Generalizado (GDA) y el Análisis Discriminante Lineal (LDA) son técnicas de clasificación ampliamente utilizadas, y cada una tiene sus propias ventajas y desventajas en función del contexto del problema. Aquí, destacaremos algunas ventajas de la Regresión Logística en comparación con LDA y GDA, desde una perspectiva más técnica y matemática.

En primer lugar, la Regresión Logística es altamente flexible en cuanto a las suposiciones que hace sobre la distribución de las características. No impone restricciones específicas en la forma en que se distribuyen los datos, lo que la hace adecuada para una amplia variedad de situaciones. En contraste, tanto LDA como GDA asumen que las características siguen una distribución gaussiana, lo que podría no ser válido en todos los casos.

Otra ventaja de la Regresión Logística es su capacidad para manejar problemas de clasificación con clases desbalanceadas. Esto se logra ajustando los umbrales de decisión y es útil en situaciones donde las clases tienen tamaños desiguales. En contraste, LDA y GDA asumen que las clases tienen la misma matriz de covarianza, lo que podría no ser apropiado para problemas desbalanceados.

Además, la Regresión Logística ofrece la ventaja de tener coeficientes interpretables en términos de log-odds. Esto significa que podemos entender cómo cada característica afecta las probabilidades de pertenecer a una clase específica, lo que puede ser valioso en la toma de decisiones. En contraste, los coeficientes en LDA y GDA se utilizan para determinar funciones de discriminación, que pueden no ser tan directamente interpretables en términos de log-odds.

Por último, la Regresión Logística permite aplicar técnicas de regularización, como la regresión logística L1 o L2, para prevenir el sobreajuste y seleccionar características relevantes. LDA y GDA no incluyen directamente términos de regularización, lo que puede hacerlos más susceptibles al sobreajuste en situaciones de alta dimensionalidad.

Clasificador Stochastic Gradient Descent (SGD)

El modelo de clasificación que utiliza SGD es una máquina de soporte vectorial (SVM) lineal o una regresión logística, dependiendo de la configuración. 

Función de Costo:
El SGDClassifier utiliza una función de costo llamada "entropía cruzada" (log loss) para medir la discrepancia entre las predicciones y las etiquetas reales. La función de costo para una muestra se define como:⁡

J(w,b)=−[ylog(h(x))+(1−y)log(1−h(x))]


- J(w,b) es la función de costo.
- y es la etiqueta real (0 para clase negativa y 1 para clase positiva).
- h(x) es la probabilidad predicha por el modelo.
  
Optimización con Descenso de Gradiente Estocástico:
El objetivo es minimizar la función de costo ajustando los pesos 

w y el sesgo 
 
b. El algoritmo de descenso de gradiente estocástico (SGD) se utiliza para actualizar los parámetros iterativamente. Para cada muestra de entrenamiento, se realiza una actualización de los parámetros de acuerdo con la siguiente regla:

w:=w−α∇J(w,b)

b:=b−α 
db
dJ(w,b)
​
Donde:
w y 

b son los parámetros a ajustar.

α es la tasa de aprendizaje que controla la velocidad de convergencia.

∇J(w,b) es el gradiente de la función de costo con respecto a w.

### svd

```python
def svd_compressor(image, order):
    """Returns the compressed image channel at the specified order"""
    
    # Create an array filled with zeros having the shape of the image
    compressed = np.zeros(image.shape)
    
    # Get the U, S and V terms (S = SIGMA)
    U, S, V = np.linalg.svd(image)
    
    # Loop over U columns (Ui), S diagonal terms (Si) and V rows (Vi) until the chosen order
    for i in range(order):
        Ui = U[:, i].reshape(-1, 1)
        Vi = V[i, :].reshape(1, -1)
        Si = S[i]
        compressed += (Ui * Si * Vi)
    
    return compressed
```

```python
rmse = []
plt.figure(figsize=(25, 10))
orders = [1,5,10,15,20,25,28]
for n in range(3):
    plt.figure(figsize=(25, 10))
    for i,order in enumerate(orders):
        
        gray_comp = svd_compressor(X_train[n].reshape(28,28), order)

        # Round the calculated values and convert them from float to int.
        gray_comp = np.around(gray_comp).astype(int)

        # Calculate RMSE
        rmse.append(np.sqrt(np.mean((X_train[n].reshape(28,28) - gray_comp)**2)))

        # Display the compressed gray image in the subplot

        plt.subplot(2, 4, i + 1)
        plt.title("Order = {} | RMSE = {:.2f}".format(order, rmse[i]))

        plt.axis('off')
        plt.imshow(gray_comp,cmap=plt.cm.gray_r,interpolation="none")
    
    plt.show()


```

##  SVD

```python
def best_svd_component_model(X_train,X_test,y_train,y_test):
    orders = [1,5,10,15,20,25,28]
    acc = []
    for k,order in enumerate(orders):
    
        X_train_2 = [svd_compressor(X_train[n].reshape(28,28),order).reshape(784,) for n in range(len(X_train))]
        X_test_2 = [svd_compressor(X_test[n].reshape(28,28),order).reshape(784,) for n in range(len(X_test))]
        logistic_model = LogisticRegression(solver='sag', multi_class='auto',max_iter=1000).fit(X_train_2,y_train)
        accuracy = accuracy_score(y_test, logistic_model.predict(X_test_2))* 100
        print(accuracy)
        acc.append(accuracy)

    plt.figure(1, figsize=(8,8))
    plt.plot(orders,acc, linewidth=2)
    plt.scatter(orders,acc)
    plt.xlabel('# order')
    plt.ylabel('Accuracy (%)')

    plt.show()
```
